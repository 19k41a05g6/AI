```python
import math
import numpy as np
x = np.random.uniform(0,100)
y = np.random.uniform(0,100)
eta = 0.01
iter_max = 200
```

iterations 1 and 2, with random x0=1 and y0=1

```python
x0=1 ; y0=1
slope_x = 6*x0
x1 = x0 - eta*slope_x
x0=x1
x2 = x1 - eta*slope_x
x2 = '{0:.2g}'.format(x2)
print("Iterations of variable x\nIteration 1: ",x1,"\nIteration 2: ", x2)
slope_y = -5*math.exp(-y0)
y1 = y0 - eta*slope_y
y0=y1
y2 = y1 - eta*slope_y
y1 = '{0:.3g}'.format(y1)
y2 = '{0:.3g}'.format(y2)
print("Iterations of variable y\nIteration 1: ",y1,"\nIteration 2: ", y2)
```

```python
iter=1
while iter <= iter_max :
  slope_x = 6*x
  steplength_x = -eta*slope_x
  x = x + steplength_x
  slope_y = -5*math.exp(-y0)
  steplength_y = -eta*slope_y
  y = y + steplength_y
  iter = iter + 1
obj = 3*x*x + 5*math.exp(-y) + 10
print("Control variables(x,y): ",round(x),",",round(y),"\nObjective Function =",round(obj))
```
