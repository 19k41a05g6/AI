```python
import numpy as np
x=np.random.uniform(-100,100)
eta=0.001
iter_max=200
```

Iteration 1 and 2 with random element x0=2

```python
x0 = 1
slope1 = 4*x0*x0*x0 + 6*x0
x1 = x0 - eta * slope1
x0 = x1
x2 = x1 - eta * slope1
print("Iteration 1 =",x1,"\nIteration 2 =",x2)
```

```python
iter = 1
while iter <= iter_max:
  slope = 4*x*x*x + 6*x
  steplength = (-eta) * slope
  x = x + steplength
  iter = iter + 1 
obj = x*x*x*x + 3*x*x + 10  
print("\nControl variable =",round(x),"\nObjective function =",round(obj),"\nGlobal minimum is at point x=",round(x),"and value of the function at this point is",round(obj))
```
